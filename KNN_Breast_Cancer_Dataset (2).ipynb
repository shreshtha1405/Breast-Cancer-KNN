# Breast Cancer Prediction using KNN

## 📌 Objective
The goal of this project is to predict whether a tumor is malignant or benign using the K-Nearest Neighbors (KNN) algorithm.

## 📊 Dataset
- Dataset: Breast Cancer Dataset
- Target Variable: Diagnosis (Malignant / Benign)

## ⚙️ Technologies Used
- Python
- Pandas
- NumPy
- Scikit-learn
- Matplotlib (optional)

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
```

```python
df.head()
```

```python
df.shape
```

## 🔧 Data Preprocessing

- Removed unnecessary columns
- Separated features (X) and target (y)
- Split dataset into training and testing sets (80-20)
- Applied StandardScaler for feature scaling

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['diagnosis'],axis=1),df['diagnosis'],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test
```

## 🤖 Model Building (KNN)

- Used KNeighborsClassifier
- Trained model on scaled training data
- Tested model on unseen test data

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
Knn=KNeighborsClassifier(n_neighbors=9)
```

```python
Knn.fit(X_train,y_train)
```

```python
y_pred=Knn.predict(X_test)
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

## 📈 Model Evaluation

- Accuracy: 96%

### Classification Report Insights:
- The model performs well for both classes (Benign and Malignant).
- Precision and recall values are high, indicating good prediction capability.
- Slightly lower recall for malignant cases (0.93) suggests that a few cancer cases are misclassified.

### Key Takeaway:
- The model is effective, but in real-world medical applications, improving recall for malignant cases is crucial.

```python
scores=[]
for i in range(1,16):
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train,y_train)
  y_pred = knn.predict(X_test)
  scores.append(accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,16),scores)
```

## 🧠 Conclusion

The KNN model performed well on the dataset and was able to classify tumors with good accuracy. Feature scaling improved the performance significantly.
